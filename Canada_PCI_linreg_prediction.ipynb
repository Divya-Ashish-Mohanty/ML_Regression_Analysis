```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
df = pd.read_csv("canada_per_capita_income.csv")
```

```python
df
```

```python
df.rename(columns = {'per capita income (US$)' : 'per_capita_income'}, inplace = True)
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per_capita_income (US$)')
plt.scatter(df.year, df.per_capita_income, color = 'blue', marker = '+')
```

```python
lin_reg.fit(df[['year']], df.per_capita_income)
```

```python
lin_reg.predict([[2020]])
```

```python
a = []
i = 2020
while i<=3000:
    a.append(i)
    i = i+1
```

```python
pred_dict = {
    "year" : a
}
```

```python
pred_df = pd.DataFrame(pred_dict)
```

```python
pred_df.to_csv('pred_year.csv', index = False)
```

```python
pred_df
```

```python
pci_pred = lin_reg.predict(pred_df)
```

```python
pred_df['per_capita_income'] = pci_pred
```

```python
pred_df
```

```python
result = pd.concat([df, pred_df], axis = 0, ignore_index = True)
```

```python
result
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel('predicted_per_capita_income(US$)')
plt.scatter(result.year, result.per_capita_income, color = 'red', marker = '+')
```
